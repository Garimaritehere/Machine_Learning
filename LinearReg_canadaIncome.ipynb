```python
# given list of independent variable-Area and dependent variable Price
#find out a model to predict Price of new given input(area)

#importing needed packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

#reading data via csv file
df=pd.read_csv("C:\datasets\canada_per_capita_income.csv")
df
```

```python
#df.year
#df.per capita income
list(df.columns)
```

```python
type(df.columns)
```

```python
df=df.rename(columns={"per capita income":"per_capita_income"})
df
```

```python
#scatter plot to visualize data
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income US($)')
plt.scatter(df.year,df.per_capita_income,color='red',marker='+')
```

```python
#preparing independent and dependent variables
X=df[['year']]
y=df.per_capita_income
#training model
reg=linear_model.LinearRegression()
reg.fit(X,y)
```

```python
#prediction
reg.predict([[2005]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#to see how my linear equation line looks
plt.xlabel('year')
plt.ylabel('per_capita_income US($)')
plt.scatter(df.year,df.per_capita_income,color='red',marker='+')
plt.plot(df.year,reg.predict(df[['year']]),color='blue')
```
